```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv('data/data_preprocessed.csv')
```

```python
df['gender_cd'] =df['gender_cd'].map({1.0: 'Мужчина', 0.0: 'Женщина'})
df.rename({'gender_cd':'Пол клиента'}, axis=1, inplace=True)
```

```python

plt.figure(figsize=(8, 6))

sns.histplot(data=df, x='age', bins=15, kde=True, hue='Пол клиента',
             palette=["#5170E2", "#D64E4E"], alpha=0.75)

plt.xlabel('', fontsize=18, labelpad=20.0)
plt.ylabel('', fontsize=18, labelpad=20.0)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.tight_layout()
plt.savefig('age_hist.png', bbox_inches='tight')
plt.show()
```
