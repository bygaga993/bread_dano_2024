```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/data_preprocessed.csv')
```

```python
IQR = df["state_talk_time_sec"].quantile(0.75) - df["state_talk_time_sec"].quantile(0.25)
high = df["state_talk_time_sec"].quantile(0.75) + 3*IQR
```

```python
df_filter = df[(df["state_talk_time_sec"] < high) & (df["state_talk_time_sec"] > 0)]
```

```python
df_combined = pd.DataFrame({
    'Original': df['state_talk_time_sec'],
    'Filtered': df_filter['state_talk_time_sec']
})

# Создаем боксплот
fig = plt.figure(figsize=(5, 8))
sns.set_style("whitegrid")  # Настройка стиля фона

# Устанавливаем цвет фона
fig.set_facecolor('#F6F6F6')
sns.boxplot(data=df_combined, color='#7C6AB2')

# Убираем подписи осей
plt.xlabel('')
plt.ylabel('')

# Показываем график
plt.title('')
plt.savefig('boxplot.png', bbox_inches='tight')
plt.show()
```
